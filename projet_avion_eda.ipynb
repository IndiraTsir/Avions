### Outliers

```python
# Create a scatter plot
plt.scatter(df_data_mini_mini.index, df_data_mini_mini['ARR_DELAY'])

# Set the title and labels
plt.title('Scatter Plot - ARR_DELAY')
plt.xlabel('Index')
plt.ylabel('ARR_DELAY')

# Display the plot
plt.show()
```

```python
# Calculate the IQR
Q1 = df_data_mini_mini['ARR_DELAY'].quantile(0.25)
Q3 = df_data_mini_mini['ARR_DELAY'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find the outliers
outliers = df_data_mini_mini[(df_data_mini_mini['ARR_DELAY'] < lower_bound) | (df_data_mini_mini['ARR_DELAY'] > upper_bound)]

# Display the outliers
print(outliers.sum)
```

## Delay by month and by airline delay
